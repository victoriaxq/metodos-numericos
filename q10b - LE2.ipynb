```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
#Solução Numérica

def dPdt(P, t):
    return [P[1], -P[1] + np.cos(t)]

P0 = [0, 0]

ts = np.arange(0, 10, 0.01)
solution = odeint(dPdt, P0, ts)
ys = solution[:, 0]

graph = plt.subplots(figsize=(6,4))

plt.plot(ts,ys)
plt.xlabel('t')
plt.ylabel('Solução Numérica y(t)')
```

```python
#Solução Particular

ypt = (1/2)*ts*np.sin(ts)
graph = plt.subplots(figsize=(6,4))
plt.plot(ypt)
plt.xlabel('t')
plt.ylabel('Solução Particular y(t)')
```

```python
#Comparando ambas as soluções

graph = plt.subplots(figsize=(6,4))
plt.plot(ypt)
plt.plot(ys)
plt.xlabel('t')
plt.show()
```
