```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def deriv(y, t):
  dydt = y * np.exp(t)#(y+4)*np.cos(t)
  return dydt

y0 = 1
time = np.linspace(0, 2, 100)
res = odeint(deriv, y0, time)

def analitica(t):
  a = np.exp(np.exp(t))*np.exp(-1)
  return a

res2 = analitica(time)
```

```python
fig = plt.figure(facecolor='pink', figsize = (10, 6))
ax = fig.add_subplot(111, facecolor='white', axisbelow=True)
ax.set_title('Questão 6: dy/dt = e^(e^t - 1)',fontsize = 12)
ax.plot(time, res, 'black', alpha=1, lw=2, label='Solução Numérica')
ax.plot(time, res2, "#d667cd", alpha=1, lw=2, ls=":", label='Solução Analítica')
ax.set_xlabel('t')
ax.set_ylabel('y(t)')
legend = ax.legend()
plt.show()
```

```python
def deriv2(y, t):
  dydt = (y+4)*np.cos(t)
  return dydt

y0 = 1
time2 = np.linspace(0, 50, 500)
res3 = odeint(deriv2, y0, time2)

def analitica2(t):
  a = np.exp(np.sin(t))*5 - 4
  return a

res4 = analitica2(time2)
```

```python
fig = plt.figure(facecolor='pink', figsize = (10, 6))
ax = fig.add_subplot(111, facecolor='white', axisbelow=True)
ax.set_title('Questão 6: dy/dt = (y + 4)*cos(t)',fontsize = 12)
ax.plot(time2, res3, 'black', alpha=1, lw=2, label='Solução Numérica')
ax.plot(time2, res4, "#d667cd", alpha=1, lw=2, ls=":", label='Solução Analítica')
ax.set_xlabel('t')
ax.set_ylabel('y(t)')
legend = ax.legend()
plt.show()
```
